## Готовые рекурсивные функции

В этих задачах дается готовая рекурсивная функция и спрашивается что-то про неё. В простейшем случае это может быть значение этой функции при каком-то значении аргумента. 

```python
#pol11
def fac(n):
    if n<1:
        return n
    if n%2==0:
        return n+3*fac(n-3)
    else:
        return 5*n+2*fac(n-5)


print(fac(31)) 
```

```python
#pol24
def F(n, s=[]):
    # print('*')
    s.append(1)
    if n >= 1:
        # print('*')
        s.append(1)
        s = F(n-1, s)
        s = F(n//3, s)
        # print('*')
        s.append(1)
    return s

s = []
s = F(280, s)
print(len(s))
```

```python
#statgrad2
def f(n):
    if n > 0 and n % 3 == 0:
        return f(n // 3)
    elif n % 3 > 0:
        return (n % 3) + f(n - (n % 3))
    else: return 0

for n in range(1, 1000):
    if f(n) == 11:
        print(n)
```
