```python
def f5(n):
    if n % 2 == 0:
        n //= 2
    else:
        n -= 1
    if n % 3 == 0:
        n //= 3
    else:
        n -= 1
    if n % 5 == 0:
        n //= 5
    else:
        n -= 1
    return n

answers = [n for n in range(1, 1000) if f5(n) == 1]
print(len(answers), answers)
```

```python
# Словарь родителей каждой вершины
d = {
     'А': '',
     'Б': 'А', 
     'В': 'АГ', 
     'Г': 'А',
     'Д': 'БВЕ',
     'Е': 'ВГ',
     'Ж': 'ЕИ',
     'И': 'ГЕ',
     'К': 'ДЛ',
     'Л': 'ДЕЖ',
     'М': 'ИЖКЛ',
     'Н': 'ИМ',
     'П': 'КМ',
     'Р': 'МН',
     'С': 'МПР'
     }

def get_label(node, d, start_node=None):
    if node == start_node or (start_node is None and not d[node]):
        return 1
    label = 0
    for parent in d[node]:
        label += get_label(parent, d, start_node)
    return label

print((get_label('Ж', d, 'А'), get_label('С', d, 'Ж')), get_label('Ж', d, 'А') * get_label('С', d, 'Ж'))
```

```python
res = dict()

def gameResult(x, y):
    if x + y >= 79: return 0
    if (x, y) in res: return res[(x, y)]

    nextCodes = [gameResult(x+1, y),
                 gameResult(x*3, y),
                 gameResult(x, y+1),
                 gameResult(x, y*3)]

    negative = [c for c in nextCodes if c <= 0]

    if negative:
        res[(x, y)] = -max(negative) + 1
    else:
        res[(x, y)] = -max(nextCodes)
    return res[(x, y)]

for s in range(1, 73):
    r = gameResult(6, s)
    if r == 2:
        print((6, s))
```

## 27А

Используем простой перебор всех возможных вариантов включить в сумму какие-то пары. Но начинаем с таких, где пар больше всего.

```python
from itertools import product
f = open('27A.txt')
n = int(f.readline())
pairs = sorted([sorted([int(x) for x in line.split()], reverse=True) for line in f \
                if int(line.split()[0]) % 2 == 1], \
               key=lambda x: sum(x), reverse=True)

m = len(pairs)
print(m)
print(*pairs, sep='\n')
max_sum_all = 0

for c in product([1, 0], repeat=m):
    # print(c)
    sum_max = sum([pairs[i][0] for i in range(m) if c[i] == 1])
    sum_min = sum([pairs[i][1] for i in range(m) if c[i] == 1])
    # print(sum_max, sum_min)
    sum_all = sum_max + sum_min
    if sum_max % 2 == 1 and sum_min % 2 == 0 and sum_all > max_sum_all:
        max_sum_all = sum_all
        print(c, sum_max, sum_min, sum_all)  

print(max_sum_all)
```

## 27B

Опять используем простой перебор возможных вариантов включить в сумму какие-то пары, но не всех вариантов, а только первых $10000$, например. Начинаем с таких, где пар больше всего. В этом случае быстро находится первая подходящая сумма, в которую входят $30108$ пар из $30110$. Остальные из пребираемых сумм оказываются меньше первой. С очень высокой вероятностью это будет правильным ответом.

```python
%%time
from itertools import product
f = open('27B.txt')
n = int(f.readline())
pairs = sorted([sorted([int(x) for x in line.split()], reverse=True) for line in f \
                if int(line.split()[0]) % 2 == 1], \
               key=lambda x: sum(x), reverse=True)
 
m = len(pairs)
print(m)
max_sum_all = 0
iterations = 0
 
for c in product([1, 0], repeat=m):
    # print(c)
    sum_max = sum([pairs[i][0] for i in range(m) if c[i] == 1])
    sum_min = sum([pairs[i][1] for i in range(m) if c[i] == 1])
    sum_all = sum_max + sum_min
    if sum_max % 2 == 1 and sum_min % 2 == 0 and sum_all > max_sum_all:
        max_sum_all = sum_all
        print(sum(c), sum_max, sum_min, sum_all)
        # print(c)
 
    if iterations > 10000:
        break
    iterations += 1 
 
print(max_sum_all)
```
