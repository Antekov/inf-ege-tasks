Обозначим через $ДЕЛ(n, m)$ утверждение «натуральное число $n$ делится без
остатка на натуральное число $m$».
Для какого наибольшего натурального числа $A$ формула
$ДЕЛ(70, A) \wedge \left(ДЕЛ(x, 28) \to \left(\neg ДЕЛ(x, A) \to \neg ДЕЛ(x, 21)\right)\right)$
тождественно истинна, то есть принимает значение $1$ при любом
натуральном $x$?

image.png

```python
def f(x, A):
    return (70 % A == 0) and ((x % 28 == 0) <= ((x % A != 0) <= (x % 21 != 0)))

for A in range(1, 100):
    A_is_good = True
    for x in range(1, 10000):
        if f(x, A) == False:
            A_is_good = False
            break
    if A_is_good:
        print(A)
```


Для какого наименьшего целого неотрицательного числа $A$ выражение

$$ (3x + 4y ≠ 60) ∨ ((A ≥ x) ∧ (A ≥ y)) $$

тождественно истинно при любых целых неотрицательных $x$ и $y$?

https://inf-ege.sdamgia.ru/problem?id=17336

В этой задаче важно начинать перебор x y именно с 0

```python
def task15():
    for a in range(0, 30):
        status = True
        for x in range(800):
            for y in range(800):
                if ((3 * x + 4 * y != 60) or ((a >= x) and (a >= y))) == 0:
                    status = False
                    break
        if status:
            print(a)
            return
task15()
```
