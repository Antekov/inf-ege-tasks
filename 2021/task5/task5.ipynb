## Посимвольное двоичное преобразование

**Задача 3**

Алгоритм получает на вход натуральное число $N > 1$ и строит по нему новое
число $R$ следующим образом:
1. Строится двоичная запись числа N.
2. Вместо последней (самой правой) двоичной цифры дважды записывается
вторая слева цифра двоичной записи.
3. Результат переводится в десятичную систему.

Пример. Дано число $N = 19$. Алгоритм работает следующим образом:
1. Двоичная запись числа $N$: $10011$.
2. Вторая слева цифра $0$, единица в конце записи заменяется на два нуля,
новая запись $100100$.
3. Результат работы алгоритма $R = 36$.

При каком наименьшем числе $N$ в результате работы алгоритма получится
$R > 92$? В ответе запишите это число в десятичной системе счисления.

> [Статград, 02-2021](../tests/)

```
def Algo(n):
    r = bin(n)[2:]
    d = r[1]
    r = r[:-1] + d + d

    s = 0
    for i in range(len(r)):
        if r[i] == '1':
            s += 2 ** (len(r) - i - 1)

    return s

for n in range(2, 93):
    if Algo(n) > 92:
        print(n, Algo(n))
```
