## Задача 3

Тимофей составляет  5 -буквенные коды из букв  Т ,  И ,  М ,  О ,  Ф ,  Е ,  Й . Буква  Т  должна входить в код ровно один раз.

Сколько различных кодов может составить Тимофей?

````
s = 'ТИМОФЕЙ'
count = 0
for c1 in s:
  for c2 in s:
    for c3 in s:
      for c4 in s:
        for c5 in s:
          word = c1 + c2 + c3 + c4 + c5
          if word.count('Т') == 1:
            count += 1
count
````

## Задача 6

Тимофей составляет $5$-буквенные коды из букв $Т$, $И$, $М$, $О$, $Ф$, $Е$, $Й$. Буква $Т$
должна входить в код не менее одного раза, а буква $Й$ – не более одного раза.

Сколько различных кодов может составить Тимофей?

В доступном наборе букв все буквы разные. Присвоим каждой букве номер из диапазона от $0$ до $6$ (всего $7$):

| Т | И | М | О | Ф | Е | Й |
|---|---|---|---|---|---|---|
|0|1|2|3|4|5|6|7|

Тогда любое слово из $5$ букв будет соответствовать списку пяти чисел из диапазона $[0; 6]$

Сделаем пять вложенных циклов `for`, в переменных которых будут перебираться значения из диапазона $[0; 6]$:

```python
for c1 in range(7):
  for c2 in range(7):
    ...
          for c5 in range(7):
            pass
```

Внутри самого глубокого цикла сформируем список из переменных этих циклов:
```python
word = [c1,c2,c3,c4,c5]
```

Остается только проверить, удовлетворяет ли полученное слово в виде списка чисел условию задачи. Если да, то увеличиваем счётчик. По условию в слове должна быть как минимум одна буква $Т$, то есть в списке количество чисел $0$ должно быть $\geq 1$. Количество букв $Й$, то есть чисел $6$ в списке должно быть $\leq 1$:

```python
if word.count(0) >= 1 and word.count(6) <= 1:
  count += 1
```

````
count = 0
for c1 in range(7):
    for c2 in range(7):
        for c3 in range(7):
            for c4 in range(7):
                for c5 in range(7):
                    word = [c1,c2,c3,c4,c5]
                    if word.count(0) >= 1 and word.count(6) <= 1:
                        count += 1
print(count)
````
