## Что такое консоль и где её взять?

Консоль (командная строка) &mdash; это текстовый интерфейс для запуска и выполнения команд и файлов с исполняемым кодом. В консоли можно текстом набрать имя команды или файла и по нажатию Enter запустить их на выполнение. 

Если команда или файл будет ожидать ввода каких-то данных от пользователя, то их можно ввести с клавиатуры в консоль. Если команда или файл должна показать какие-то результаты пользователю, то она тоже может вывести нужные значения в консоль.

После окончания выполнения консоль снова будет ждать запуска следующей команды или файла.

Консоль можно получить с помощью:

- меню Пуск &rarr; Служебные &rarr; Командная строка
- поиска в проводнике файла `cmd.exe` и его запуска
- редактора кода с интегрированной консолью, например, в [Visual Studio Code](https://code.visualstudio.com/) нажав <code>Ctrl + `</code>





image.png

Напомним, что *полное имя файла* состоит из *пути* и *имени файла*. Путь к файлу &mdash; это последовательность вложенных друг в друга папок относительно корневой папки (полный путь) или текущей папки (относительный путь), внутри самой глубокой из которых находится файл.

Последняя строка на картинке выше (`C:\Users\Antekov>`) называется строкой приглашения. В ней мы видим путь к текущей папке. Именно в ней в первую очередь будут искаться файлы, которые мы захотим запустить из консоли (по имени файла, т.е. без указания пути к файлу). На других компьютерах текущаяя папка при запуске консоли, скорее всего, будет другой.


Часть имени файла, идущая после последней точки в имени, называется *расширением*. Саму последнюю точку можно тоже относить к расширению.

Из консоли можно запустить на выполнение только файл с исполняемым кодом (исполняемый файл). Они, как правило, имеют расширение `.exe`

## Как запустить программу на Python из консоли?

> Скорее всего, вам не придется запускать на экзамене файлы таким образом.

Для запуска на выполнение программы, хранящейся в файле `example.py` необходимо в консоли выполнить команду

```bash
C:\Users\Antekov\>python example.py
```

то есть написать имя интерпретатора и имя файла с программой.

Если файл с программой лежит в текущей папке, то достаточно указать только имя файла без пути. Если не лежит, то указываем полное имя файла. Для запуска файла `C:/Python/Workspace/example.py` команда будет выглядеть так:

```bash
C:\Users\Antekov\>python C:/Python/Workspace/example.py
```

Можно указать и относительное имя файла: если текущая папка `C:/Users/Antekov/`, то чтобы добраться до вышеупомянутого файла необходимо подняться вверх на два уровня папок (это делает часть относительного пути из двух комбинаций `../`), а затем спуститься в нужную папку (это делает оставшаяся часть относительного пути):

```bash
C:\Users\Antekov\>python ../../Python/Workspace/example.py
```

Если в полном имени файла присутствуют пробелы, то необходимо взять его в двойные кавычки:

```bash
C:\Users\Antekov\>python "C:/Python/Cool Workspace/example.py"
```

При необходимости может понадобится указать полный путь не только к файлу с программой, но и к файлу интерпретатора:

```bash
C:\Users\Antekov\>C:/Python/Python39/python.exe C:/Python/Workspace/example.py
```

## Что представляет собой программа на Python?

Программа на Python &mdash; это, как правило, тектстовый файл с расширением `*.py`, содержащий команды, которые сможет выполнить интерпретатор `python.exe`.

Команды могут содержать:
- синтаксические конструкции (циклы, условный оператор, ...)
- выражения (присвоение значений переменным, арифметические выражения, ...)

Если какой-то набор команд требуется использовать неоднократно, то ему можно присвоить имя и затем, в том месте программы, где понадобится его выполнить, это можно будет сделать по имени. Такой блок кода с именем называется *функцией*.

В Python есть много уже готовых функций, которые можно запускать сразу просто по имени из своей программы. Первыми функциями, с которыми мы встретимся, будут:
```python
print(...)  # Выводит в консоль то, что ей передано внутри круглых скобок
input()     # Запрашивает у пользователя последовательность символов, вводимых с клавиатуры 
```

Еще больше готовых функций есть в дополнительных модулях, которые устанавливаются вместе с интерпретатором, но доступными они становятся только после команды импорта, например:
```python
import math  # Импортировать модуль math, содержащий математические функции типа sin, cos и др.
```

Любую простую программу можно представить состоящей из трёх частей, выполняющих следующие действия:
- ввод данных
- обработка данных
- вывод результата

В более сложных программах эти три части могут встречаться несколько раз или вообще представлять собой бесконечный цикл, заканчивающийся только при завершении программы пользователем.

В некоторых бесполезных программах какие-то из этих трёх частей могут отсутствовать. В некоторых более полезных программах эти три части могут быть написаны так, что их трудно разделить между собой на отдельные строки кода, например:
```python
print('Hello,' + input())
```

### Целые числа: `int`

Три основных способа получения переменных, содежащих целые числа:

- присвоить переменной последовательность символов цифр
```python
a = 2100
```
- преобразовать строку в число с помощью функции `int()`:
```python
a = int('2100')
```
- преобразовать строку, содержащую запись числа не в десятичной системе, в число с помощью той же функции `int()`:
```python
a = int('4064', 8) # 4064 в восмеричной = 2100 в десятичной
```

Продемонстрируем примеры:

````python
a = 2100
print(a)
````

````python
a = int('2100')
print(a)
````

````python
a = int('4064', 8)  # 4064 в восмеричной = 2100 в десятичной
print(a)
````

### Списки: `list`


Список &mdash; это контейнер, содержащий некоторые значения (элементы списка), каждому из которых выделено свое отдельное пронумерованное место. Номер места называется индексом элемента. Индекс самого первого элемента равен $0$, индексы следующих &mdash; целые числа с шагом $1$.

Применяется в том случае, когда нужно сохранить набор значений и их взаимное расположение друг относительно друга. Чтобы иметь возможность затем обратиться к элементам по их номеру. Значения в списке могут повторяться на разных местах.

Примеры задач, в которых удобно использовать список:
- найти сумму длинной последовательности чисел
- из набора пар чисел найти пару с минимальной разностью
- найти пять самых дорогих товаров из списка цен
- найти такую пару чисел из большого набора, чтобы их произведение делилось на 3
- отсортировать набор значений, которые можно сравнивать между собой

Во всех этих примерах мы помещаем входные данные в список и затем выполняем те или иные манипуляции со всем списком или его отдельными элементами.

Все элементы списка являются, как правило, значениями одного типа. Например, список целых чисел, список строк.

Получить переменную, содержащую пустой список можно с помощью функции-конструктора `list()` или c помощью `[]`

```python
b = []      # Получаем пустой список b = []
```

- получить новый список из другого списка или контейнера, умеющего выдвать по одному значению из некоторого набора, с помощью конструкции, называемой *генератором*:
```python
a = [1, 2, 3, 4, 5]
b = [x**2 for x in a] # Получаем список b = [1, 4, 9, 16, 25]
```

Все элементы множества тоже являются, как правило, значениями одного типа. Например, множество целых чисел, множество строк.

Применятеся в том случае, если необходимо сохранить набор уникальных значений без какого-либо взаимного расположения между значениями. Как в мешке. Чтобы иметь возможность проверить, есть ли такое значение в контейнере или нет. Или вытряхнуть из мешка все значения по одному как получится.

Пустое множество, которое потом можно наполнить значениями, можно получить с помощью функции `set()`:

```python
a = [1, 2, 3, 4, 5]
b = {x**2 for x in a}  # Получаем множество b = {1, 4, 9, 16, 25}
```

В большинстве случаев требуется присвоить переменной какое-то значение, которое должен определить пользователь уже после запуска программы. Часть кода программы, которая делает это, и является кодом ввода данных. 

Вот два способа ввода данных, которые мы будем использовать:
- ввод значений с клавиатуры
- чтение значений из файла (рассмотрим его позднее)

Необходимо представлять, что с клавиатуры можно получить от пользователя только последовательность символов. Никаких чисел, картинок, видео, линий, точек, слонов и прочего. Только символы! А вот затем полученные символы можно перобразовать уже во что-то другое. Например, в целое число, если среди введенных символов есть только символы цифр и никаких больше. 

Итак, читаем строку:
```python

```

Чтобы её не потерять, сохраняем в переменную:
```python
s = input()
```

Если входными данными должна быть строка символов, то ввод данных закончен. Если же нет, то тогда нужно дополнительно обработать эту строку, чтобы извлечь из нее значения в нужном виде.

Если в строке должно быть одно число, то передаём строку соответствующей функции преобразования:

Без использования промежуточной переменной `s` можно обойтись, поэтому, как правило, ввод целого числа с клавиатуры выглядит так:

А ввод вещественного числа &mdash; так:

## Условный оператор

Примеры условий:

```python
# Просто константа `Истина`
True

# `Истина`, если значение переменной `a` строго больше десяти, иначе `Ложь`
a > 10

# `Истина`, если при возведении в квадрат значения перменной `n` окажется 
#  строго меньше числа 1017, иначе `Ложь`
n * n < 1017

# `Истина`, если значение перменной `n` окажется чётным числом, иначе `Ложь`
n % 2 != 0

# `Истина`, если число 3 содержится в множестве чисел {2, 3, 7, 9}, иначе `Ложь`
3 in {2, 3, 7, 9}

# `Истина`, если строка перед `not in` содержится в строке,
#  которая стоит после оператора `not in`. Иначе - `Ложь`
'xyz' not in 'На заботе написано слово mir' 

# `Истина`, если строка, прочитанная с клавиатуры будет равна `Hello`,
#  иначе - `Ложь`
input() == 'Hello'
```

Условия используются в основном в условном операторе (кто бы мог подумать!). Вот его формат:

```python
if условие:
    действия, если условие истинно
```
Обратите внимание, что после условия ставится символ двоеточия `:`

Здесь мы впервые сталкиваемся с понятием блок кода. Если при выполнении условия нужно выполнить несколько действий, то необходимо иметь возможность задать границы блока кода, который должен выполнять эти действия, чтобы отделить его от остальных действий, выполняемых в программе далее. 

Во многих языках (C, C++, Java, JavaScript, ...) для обозначения блока кода используют фигурные скобки, например:
```cpp
if (a > 10) 
{
    a = a - 10;
    b = a * a;
}
c = a + b;
```
В этом примере при выполнении условия `a > 10` выполнятся команды `a = a - 10` и
`b = a * a`. После этого в любом случае будет выполнена команда `c = a + b`. То есть команды изменения переменных `a` и `b` образуют блок кода, который надо либо выполнить целиком, либо не выполнять совсем. Для задания границ этого блока как раз используются `{` и `}`.

В Python пошли другим путем. Начало блока обозначается символом двоеточия `:` Следующая строка должна иметь больший отступ от левого края. Все последующие строки с таким же (или еще большим) отступом будут считаться относящимися к этому блоку кода. 

Как только следующая строка венется к такому же отступу, как строка с заголовком `if ...`, то она уже не будет относиться к блоку кода этого оператора `if` 

```python
if условие:
    действие 1  # Блок внутри оператора if
    действие 2  # Блок внутри оператора if
    действие 3  # Блок внутри оператора if

действие 4  # Этот код уже не отностися к блоку внутри оператора if
```

Чтобы выполнить действия, необходимые в случае, если условие оператора `if` ложно, добавляется слово `else:`, после которого идёт блок кода с этими действиями:

```python
if условие:
    действия, если условие истинно
else:
    действия, если условие ложно
```

Также можно использовать `elif`, чтобы добавить блоки кода, которые должны выполняться при истинности других условий:
```python
if условие1:
    действия, если условие1 истинно
elif условие2:
    действия, если условие2 истинно (а условие1 - ложно)
elif условие3:
    действия, если условие3 истинно (а условие1 и условие2 - ложны)
else:
    действия, если все условия ложны
```

Можно вкладывать друг в друга

## Цикл `while`

В отличии от условного оператора, который проверяет условие и выполняет действия однократно, цикл while будет повторять проверку условия и выполнение действий до тех пор, пока однажды условие не станет ложным.

```python
while условие:
    действия, если условие истинно
```

После того, как условие станет ложным, смогут наконец-то выполниться действия, идущие после цикла:

```python
while солнце_пока_светит?():
    начался_новый_день()
    выполнить_дневные_задачи()
    новый_день_закончился()
    преждать_ночь()

print('Солнце погасло навсегда. Хотя, кому я это вывожу?')
```

## Цикл `for`

```python
for переменная in контейнер:
    действия для очередного элемента из контейнера,
    помещенного в переменную
```

Если контейнер умеет выдавать не по одному значению за раз, а по несколько, то подхватывать их в цикле можно таким же количеством переменных перед словом `in`. 

Если для такого контейнера поставить одну переменную, то она будет кортежем 

## Списки и генераторы

Некоторые списки используются очень часто

````python
a = range(10)
a
````

````python
b = list(range(10))
b
````

````python
print(b[3])
print(a[3])
````

````python
p = []
for i in range(1, 10):
  p.append(i**2)
print(p)

p = [i**2 for i in range(1, 10)]
print(p)
````

````python
p = []
for i in range(1, 10):
  if i != 5:
    p.append(i**2)
print(p)

p = [i**2 for i in range(1, 10) if i != 5]
print(p)
````
