## Запускаем, а он не запускается

Допустим, что вы поставили Python на свой компьютер в каталог `C:/Python/Python39`. Открываете любую доступную консоль и вводите команду:
```bash
python
```
Но в ответ получаете сообщение примерно такого типа:

```bash
"python" не является внутренней или внешней
командой, исполняемой программой или пакетным файлом.
```

Каким образом можно попробовать исправить ситуацию, чтобы получить возможность выполнить программу на Python на этом компьютере?

Выберите один или несколько ответов:

- Проверить, что путь к файлу python.exe присутствует в списке путей системной переменной PATH и добавить его при отсутствии

- Найти другой компьютер, на котором кто-то уже решил эту задачу

- Набрать в консоли полный путь: `C:/Python/Python39/python`

- Зайти под учётной записью с правами администратора и повторить команду из задачи

- Сменить текущий каталог на тот, в который устанавливался Python

- Переустановить Python таким же образом, как ставили его ранее
