### Положительный рост

В приведенном ниже коде печатается строка `'000011111'`
```python
s = '0' * 4 + '1' * 5
print(s)
```

Здесь, помимо умножения строк на числа мы имспользуем операцию сложения двух строк, обозначаемую знаком `+`. Операция умножения имеет более высокий приоритет по сравнению с операцией сложения, поэтому при вычислении выражения сначала получаются две строки `'0000'` и `'1111'`, а затем они складываются в строку `'00001111'`

Превратите этот код в печатающий строку из $40$ символов `0` и затем $50$ символов `1`

````python
s = '0' * 4 + '1' * 5
print(s)
````
