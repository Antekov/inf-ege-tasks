В этой саге будет незримо присутствовать два действующих лица: *программист* и *пользователь*.

*Программист* - пишет программу на языке Python, сохраняет ее нажатием `Ctrl+S` в своём любимом редакторе кода, называемом IDE (интегрированная среда разработки). Если повезёт. Если не повезёт - пишет код в любом текстовом редакторе начиная от стандартного Блокнота в Windows и не заканчивая MS Word или даже Excel. Обычно программисту везёт.

*Пользователь* - получает программу от программиста, запускает ее, передавая ей входные данные с клавитуры или из файла, и получает правильный результат. Если повезет. Если не повезет, то вспоминает программиста..., в смысле, про программиста, и вежливо просит его доработать программу. Обычно пользователю *не* везёт.

А порой программист и пользователь оказываются одним и тем же физическим лицом. ЕГЭ &mdash; это именно такая пора.

## Что представляет собой программа на Python?

Программа на Python - это, как правило, тектстовый файл с расширением `*.py`, содержащая команды, которые сможет выполнить интерпретатор `python.exe`.

Команды могут содержать:
- синтаксические конструкции (циклы, условный оператор, ...)
- выражения (присвоение значений переменным, арифметические выражения, ...)

Если какой-то набор команд требуется использовать неоднократно, то ему можно присвоить имя и затем, в том месте программы, где понадобится его выполнить, это можно будет сделать по имени. Такой блок кода с именем называется функцией.

В Python есть много уже готовых функций, которые можно запускать сразу просто по имени из своей программы. Первыми функциями, скоторыми мы встретимся будут:
```python
print(...) # Выводит в консоль то, что ей передано внутри круглых скобок
input()    # Запрашивает у пользователя последовательность символов, вводимых с клавиатуры 
```

Еще больше готовых функций есть в дополнительных модулях, которые устанавливаются вместе с интерпретатором, но доступными они становятся только после команды импорта, например:
```python
import math
```

Любую простую программу можно представить состоящей из трёх частей:
- ввод данных
- обработка данных
- вывод результата

В более сложных программах эти три части могут встречаться несколько раз или вообще представлять собой бесконечный цикл, заканчивающийся только при завершении программы пользователем.

В некоторых бесполезных программах какие-то из этих трёх частей могут отсутствовать. В некоторых более полезных программах эти три части могут быть написаны так, что их трудно разделить между собой на отдельные строки кода, например:
```python
print('Hello,' + input())
```

````python
a = range(10)
a
````

````python
print(b[3])
print(a[3])
````

````python
p = []
for i in range(1, 10):
  p.append(i**2)
print(p)

p = [i**2 for i in range(1, 10)]
print(p)
````

````python
p = []
for i in range(1, 10):
  if i != 5:
    p.append(i**2)
print(p)

p = [i**2 for i in range(1, 10) if i != 5]
print(p)
````
