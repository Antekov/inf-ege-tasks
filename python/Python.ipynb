## Переменные, типы данных



Переменная - это имя кусочка памяти, в котором хранится значение (данные) некоторго типа.

Когда мы присваиваем значение переменной, то она обретает значение и тип, к котрому отностися присвоенное значение:



```python
a = 100 # присвоить перменной с именем a значение, содержащее целое число 100
```

При присвоении переменной значения другого типа, ее тип меняется.

Узнать тип переменной можно с помощью функции `type()`

Вывести значение переменой можно с помощью функции `print()`

```python
type(a)
```

```python
print(a)
```

## Списки и генераторы

```python
p = []
for i in range(1, 10):
  p.append(i**2)
print(p)

p = [i**2 for i in range(1, 10)]
print(p)
```

```python
p = []
for i in range(1, 10):
  if i != 5:
    p.append(i**2)
print(p)

p = [i**2 for i in range(1, 10) if i != 5]
print(p)
```
